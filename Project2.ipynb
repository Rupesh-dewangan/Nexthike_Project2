```python
# Replacing the column attribute with season name

def change_season(x):
    if x==1:
        return "Spring"
    elif x==2:
        return "Summer"
    elif x==3:
        return "Fall"
    else:
        return "Winter"

datasetfinal['season'] = datasetfinal['season'].apply(change_season)
```

```python
datasetfinal
```

```python
# Checking the unique values of weathersit column. 
datasetfinal['weathersit'].unique()
```

```python
# Changing the values of weathersit column  as name of weater typer like 1-Clear,2-Mist,3-light Snow,Heavy Rain

def change_weather(x):
    if x == 1:
        return 'clear'
    elif x == 2:
        return 'mist'
    elif x == 3:
        return 'light snow'
    else:
        return 'heavy rain'

datasetfinal['weathersit'] = datasetfinal['weathersit'].apply(change_weather)
datasetfinal.head()
```

# Insights from data

```python
# Checking users count weather wise 
count_user_weather = datasetfinal.groupby(by=["weathersit"]).agg({
    'dteday':'count',
    "casual": "sum",
    "registered": "sum",
    "cnt":"sum"
})
count_user_weather
```

```python
ax = sns.barplot(data = count_user_weather, x = 'weathersit', y = 'cnt')
plt.ylabel("Count of User")
plt.xlabel("Weather Type")
plt.title("Total Count of users on different weather type")
for i in ax.containers:
    ax.bar_label(i,)
plt.show()
```

```python
ax = count_user_weather.plot(kind ='bar',y=['registered','casual'])
for i in ax.containers:
    ax.bar_label(i,)
plt.ylabel("Count of User")
plt.xlabel("Weather Type")
plt.title("Number of casual user and registered user based on weather")
plt.show()
```

```python
# Checking the unique values of weekday column. 
datasetfinal['weekday'].unique()

```

```python
# adding new column in dataset as weekday for day name like 0-Monday , 1-Tuesday, 2-Wednesday,3-Thurday,4-Friday,5-Saturday,6-Sunday

def change_dayname(x):
    if x == 0:
        return 'Monday'
    elif x == 1:
        return 'Tuesday'
    elif x == 2:
        return 'Wednesday'
    elif x == 3:
        return 'Thursday'
    elif x == 4:
        return 'Friday'
    elif x == 5:
        return 'Saturday'
    else:
        return 'Sunday'

datasetfinal['dayname'] = datasetfinal['weekday'].apply(change_dayname)
datasetfinal.head()
```

```python
# number of casual users on days 
sum_casualuser = datasetfinal.groupby(['weekday','dayname']).casual.sum().reset_index()

sum_casualuser
```

```python
ax = sns.barplot(data=sum_casualuser,x='dayname',y='casual')
for i in ax.containers:
    ax.bar_label(i,)
plt.ylabel("Count of User")
plt.xlabel("Days")
plt.title("Count of Casual users on weekdays")
plt.show()
```

```python
#Number of registersd user on day
sum_registered = datasetfinal.groupby(['weekday','dayname']).registered.sum().reset_index()
sum_registered
```

```python
ax = sns.barplot(data=sum_registered,x='dayname',y='registered')
for i in ax.containers:
    ax.bar_label(i,)
plt.ylabel("Count of User")
plt.xlabel("Days")
plt.title("Count of Registered users on weekdays")
plt.show()
```

```python
daily_user = pd.merge(left=sum_casualuser,right=sum_registered,how='left',left_on='dayname',right_on='dayname')
daily_user
```

```python
daily_user_type = daily_user.melt(id_vars=['dayname'],value_vars=['casual','registered'] , var_name='user_type', value_name='user_count')
daily_user_type
```

```python
#sns.pointplot(x= 'dayname',y='user_count',data=daily_user_type)
plt.figure(figsize=(12,6))
ax = sns.barplot(data=daily_user_type, x='dayname', y='user_count', hue='user_type')
for i in ax.containers:
    ax.bar_label(i,)

plt.title('Comparision for registered and casual user based on day')
plt.ylabel('Total User')
plt.xlabel('Day')
plt.show()
```

```python
count_on_holiday = datasetfinal.groupby('holiday').cnt.sum().reset_index()
count_on_holiday= count_on_holiday.replace({False:'Non Holiday', True:'Holiday'})
count_on_holiday
```

```python
# hour_count  = datasetfinal.groupby('hr').cnt.sum().reset_index()
# hour_count

a1 = datasetfinal.groupby('hr').casual.sum().reset_index()
a2 = datasetfinal.groupby('hr').registered.sum().reset_index()

hour_count1 = pd.merge(left=a1,right=a2,how='left',left_on='hr',right_on='hr')
hour_count1


hour_count1_type = hour_count1.melt(id_vars=['hr'],value_vars=['casual','registered'] , var_name='user_type', value_name='user_count')
hour_count1_type
```

```python
fig, ax = plt.subplots(figsize=(20,5))
sns.pointplot(data=hour_count1_type, x='hr', y='user_count', hue='user_type',  errorbar=None, ax=ax)
ax.set(title='count of user based on Time')
ax.set_ylabel('Total User')
ax.set_xlabel('Hour')
plt.show()
```

```python
user_count_perday  = datasetfinal.groupby('dteday').cnt.sum().reset_index()
user_count_perday


b1 = datasetfinal.groupby('dteday').casual.sum().reset_index()
b2 = datasetfinal.groupby('dteday').registered.sum().reset_index()

user_count_perday1 = pd.merge(left=b1,right=b2,how='left',left_on='dteday',right_on='dteday')

user_count_perday1['dteday'] = user_count_perday1['dteday'].dt.strftime('%d/%m/%Y')

user_count_perday1 = user_count_perday1.melt(id_vars=['dteday'],value_vars=['casual','registered'] , var_name='user_type', value_name='user_count')
user_count_perday1
```

```python
fig, ax = plt.subplots(figsize=(25,8))
sns.pointplot(data=user_count_perday1, y='user_count', x='dteday', hue='user_type',  errorbar=None, ax=ax)
ax.set(title='Count of users Based on date')

ax.set_ylabel('Total User')
ax.set_xlabel('dates')
ax.set_xticks(user_count_perday1['dteday'])
ax.set_xticklabels(user_count_perday1['dteday'], rotation=45)
plt.show()
```
